#### 3. 在Python中，函数定义时使用可变对象作为默认参数值可能会导致什么问题？

````python
def add_num(num, nums=[]):
    nums.append(num)
    return nums

print(add_num(1))
print(add_num(2))
print(add_num(3, []))
print(add_num(4))
print(add_num(5, []))

# 默认列表A在def定义时只创建一次,所有无参数调用函数都共享默认列表A,修改会被保留
# 使用传入的新列表[]（每次传都是一个全新的列表对象）,和默认列表A无关系
````
