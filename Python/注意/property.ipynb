# 装饰器的作用

## 1.创建只读属性

```python
class Person:
    def __init__(self, name):
        self.name = name
        self.__age = 18

    @property
    def age(self):
        return self.__age

p = Person('小明')
print(p.age)

try:
    p.age = 20
except AttributeError as e:
    print(f'错误：出现{e}，只读属性不能被修改')
```

## 2.简化调用

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius  # 半径
    def area(self):
        return 3.14 * self.radius ** 2
c = Circle(20)
print(c.area) # <bound method Circle.area of <__main__.Circle object at 0x0000029770EC6000>>

# 而
class Circle:
    def __init__(self, radius):
        self.radius = radius  # 半径
    @property
    def area(self):
        return 3.14 * self.radius ** 2
c = Circle(20)
print(c.area) # 把方法当作属性调用
```

## 3.属性赋值校验
@property 定义读方法  
@xxx.setter 定义写方法  
配合 `@属性名.setter`，给属性赋值时自动校验数据，避免无效值

```python
class Student:
    def __init__(self):
        self.__score = 0

    # 第一步：定义读属性
    @property
    def score(self):
        return self.__score

    # 第二步：定义写属性 + 数据校验
    @score.setter
    def score(self, value):
        if not 0 <= value <= 100: # 检查
            raise ValueError("分数必须在0-100之间！")
        self.__score = value

s = Student()
print(s.score)
s.score = 90
print(s.score)
try:
    s.score = 150
except ValueError as e:
    print(f'错误：{e}')
```
