# yield生成器函数

```python
def number_generator():
    print("执行第一段代码")
    yield 1          # 第一次暂停，返回1
    print("执行第二段代码")
    yield 2          # 第二次暂停，返回2
    print("执行第三段代码")
    yield 3          # 第三次暂停，返回3

# 1. 调用函数：仅创建生成器对象，**不执行任何代码**
gen = number_generator()
print("生成器对象：", gen)

# 2. 第一次取值：函数开始执行，遇到yield暂停
print(next(gen))  # 输出：执行第一段代码 → 1

# 3. 第二次取值：从暂停处继续执行，遇到yield再暂停
print(next(gen))  # 输出：执行第二段代码 → 2

# 4. 第三次取值：再次继续执行，遇到yield暂停
print(next(gen))  # 输出：执行第三段代码 → 3

try:
    print(next(gen))  # 会报错：StopIteration
except StopIteration as e:
    print(f'错误：{e} StopIteration')
```
