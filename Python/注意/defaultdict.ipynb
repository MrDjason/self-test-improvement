# defaultdict

defaultdict是Python 自带的增强版字典，提前设定一个默认值

```python
import numpy as np
X = np.zeros((3,3))
print(X)
v = {}
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        state = (i, j)
        v[state] = 0
print(v)
```

```python
# defaultdict
import numpy as np
from collections import defaultdict
V = defaultdict(lambda: 0)
print(V)
print(V[100,100])
```
