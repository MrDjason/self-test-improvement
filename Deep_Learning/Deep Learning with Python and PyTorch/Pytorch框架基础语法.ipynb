# PyTorch框架基础语法



```python
class Net(nn.Module):
    def __init__(self,):
        super(Net, self).__init__()
        self.layer1 = nn.Sequential(nn.Linear(in_dim, n_hidden_1), nn.BatchNorm1d(n_hidden_1))
    def forward(self):
        x = F.relu(self.layer1(x))
```

- class Net(nn.Module):`nn.Module`是Pytorch所有模型必须继承的父类
- \_\_init\_\_、forward:类的两个固定函数
- super(模型类名, self).\_\_init\_\_():
- nn.Sequential:把多个网络层打包在一起，简化代码
- nn.Linear(输入维度, 输出维度):基础网络层，用于数据维度的转换
- nn.BatchNorm1d(维度):辅助训练，让模型训练更快、更稳定
- F.relu(数据):给网络层添加非线性运算
