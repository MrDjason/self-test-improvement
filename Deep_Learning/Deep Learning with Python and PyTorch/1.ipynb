## 1.2 获取元素

```python
# 创建数据
np.random.seed(2019)
np11 = np.random.random([101])

# 获取数据
print(np11[3])    # 截取一个
print(np11[3:6])  # 截取一段
print(np11[1:6:2])# 间隔取数
print(np11[::-2]) # 倒叙

np12 = np.arange(25).reshape([5,5])
print(np12)
print(np12[1:3,1:3]) # 截取区域内数据
print(np12[(np12>3) & (np12<10)]) # 截取一个多维数组中数值在一个值域之中的数据

print(np12[[1,2]]) # 读取指定两行，或者print(np12[1:3,:])
print(np12[:,1:3]) # 读取2，3列
```

## 1.4 数组变形

#### 1.4.1 更改数组形状
|函数|描述|
|:---:|:---:|
|arr.reshape|重新将向量arr进行维度变换，不修改向量本身|
|arr.resize|重新将向量arr进行维度变化，修改向量本身|
|arr.T|对向量arr进行转置|
|arr.flatten|对向量arr进行展平，即将多维数组变为1为数组，不会产生原数组的副本|
|arr.squeeze|只能对维数为1的维度降维。对多维数组使用时不会报错，但是不会产生任何影响|
|arr.transpose|对高维矩阵进行轴对换|

**1）reshape**不修改向量本身  
可以只指定行或列进行变换，其余用-1代替

```python
arr = np.arange(10)
print(arr)
print(arr.reshape(5,2))
print(arr.reshape(5,-1))
print(arr)
```

**2）resize**修改向量本身

```python
arr = np.arange(10)
print(arr)
arr.resize(5,2)
print(arr)
```

**3）T**转置

```python
arr = np.arange(10).reshape(5,2)
print(arr)
print(arr.T)
```

**4)ravel**向量展平
分为行展开列展开

```python
arr = np.arange(6).reshape(2, -1)
print(arr)
print('按照列展开优先', arr.ravel('F'), sep='\n')
print('按照行展开优先', arr.ravel(), sep='\n')
```

**5）flatten**矩阵转换为向量

```python
arr = np.floor(10*np.random.random((3, 4)))
print(arr)
print(arr.flatten())
```

**6）squeeze**降维，将矩阵中含1的维度去掉

```python
arr = np.arange(3).reshape(3, 1)
print(arr.shape)
print(arr)
print(arr.squeeze().shape)
print(arr.squeeze())
```

**7）transpose**对高维矩阵进行轴对换  
transpose(1,2,0) → 将矩阵(0, 1, 2)维度变换为(1, 2, 0)

```python
arr = np.arange(24).reshape(2,3,4)
print(arr.shape)
print(arr.transpose(1,2,0).shape)
```

#### 1.4.2 合并数组
|函数|描述|
|:---:|:---:|
|np.append|内存占用大|
|np.concatenate|没有内存问题|
|np.stack|沿着新的轴加入一系列数组|
|np.hstack|堆栈数组垂直顺序（行）|
|np.vstack|堆栈数组垂直顺序（列）|
|np.dstack|堆栈数组按顺序深入|
|np.vsplit|将数组分解成垂直的多个子数组的列表|

**1）append**

```python
# 合并一维数组
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.append(a, b)
print(c)
# 合并多维数组
a = np.arange(4).reshape(2, 2)
b = np.arange(4).reshape(2, 2)
# 按行合并
c = np.append(a, b, axis=0)
print(c)
# 按列合并
d = np.append(a, b, axis=1)
print(d)
```

**2）concatenate**沿指定轴连接数组或者矩阵

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
# 按行合并
c = np.concatenate((a, b), axis=0)
print(c)
# 按列合并
d = np.concatenate((a, b.T), axis=1)
print(d)
```

**3）stack**沿指定轴堆叠数组或矩阵

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.stack((a, b), axis=0))
```

### 1.5 批量处理
大数据拆成多个批次：  
- 1）得到数据集
- 2）随机打乱数据
- 3）定义批次大小
- 4）批处理数据集

```python
# 生成10000个形状2×3的矩阵
data_train = np.random.randn(10000, 2, 3)
print(data_train.shape)

# 打乱
np.random.shuffle(data_train)

# 定义批次大小
batch_size = 100

# 进行批处理
for i in range(0, len(data_train), batch_size):
    x_batch_sum = np.sum(data_train[i:i+batch_size])
    print(f'第{i}批次，该批次的数据之和：{x_batch_sum}')
```
