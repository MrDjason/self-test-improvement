```python
# 缺失值处理
import pandas as pd
import numpy as np
s = pd.Series([1, 2, np.nan, None, pd.NA]) # nan: not a number
print(s)
# 勘察是否是缺失值
print(s.isna())
print(s.isnull())
df = pd.DataFrame([[1, pd.NA, 2], [2, 3, 5], [None, 4, 6]])
print(df)
print(df.isna())
print(df.isnull())
print('第0列:1,第1列1,第2列0','\n',df.isna().sum())
print(s.isna().sum()) # 查看缺失值个数
print('-'* 20)
# 剔除缺失值
print(s)
print(s.dropna())
print()
print(df)
print()
print(df.dropna()) # 剔除一整条记录  
print()
print(df.dropna(how='all')) # 如果所有的值都是缺失值，删除这一行
print()
print(df.dropna(thresh=2)) # 如果至少有n个值不是缺失值，就保留
```
