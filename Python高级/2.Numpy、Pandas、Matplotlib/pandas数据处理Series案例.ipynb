# Series案例

创建一个包含10名学生数学成绩的Series,成绩范围在50-100之间。
计算平均分、最高分、最低分,并找出高于平均分的学生人数。

```python
import numpy as np
import pandas as pd
np.random.seed(42)
values = np.random.randint(50, 101, 10)
print(values)
```

```python
indexs = []
for i in range(1,11):
    indexs.append('学生'+str(i))

scores = pd.Series(values, indexs)

print(scores)
print('平均分：', scores.mean())
print('最高分：', scores.max())
print('最低分：', scores.min())
```

```python
# 高于平均分的学生人数
mean = scores.mean()
print(scores[scores>mean])
# 两种方法
print(len(scores[scores>mean]))
print(scores[scores>mean].count())
```

给定某城市一周每天的最高温度Series,完成以下任务：  
找出温度超过30度的天数  
计算平均温度  
将温度从高到低排序  
找出温度变化最大的两天  

```python
import numpy as np
import pandas as pd
temperatures =  pd.Series([28,31,29,32,30,27,33],
                index=['周一','周二','周三','周四','周五','周六','周日'])


print('温度大于30度的天数：', len(temperatures[temperatures>30]))
print('平均温度：', temperatures.mean())

t2 = temperatures.sort_values(ascending=False)
print(t2)

t3 = temperatures.diff().abs()


print('温度变化最大的两天：',*t3.sort_values(ascending=False).keys()[:2].tolist())
```

给定某股票连续10个交易日的收盘价Series:  
计算每日收益率(当日收盘价/前日收盘价-1)  
找出收益率最高和最低的日期  
计算波动率(收益率的标准差)  

```python
import pandas as pd
import numpy as np
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1,
109.3, 110.2], index=pd.date_range('2023-01-01', periods=10))
print(prices)
```

```python
a = prices.pct_change() # percent 103.5/102.3 - 1
print(a)
# 收益率最高的日期
print(a.idxmax())
# 收益率最低的日期
print(a.idxmin())
# 波动率
print(a.std())
```

某产品过去12个月的销售量Series:  
计算季度平均销量(每3个月为一个季度)  
找出销量最高的月份  
计算月环比增长率  
找出连续增长超过2个月的月份  

```python
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210,
220],index=pd.date_range('2022-01-01', periods=12, freq='MS'))
print(sales)
```

```python
# 季度平均销量
print('季度平均销量：',sales.resample('QS').mean(),sep='\n') # Quarter Start
print('年平均销量：',sales.resample('YS').mean()) # 按年 重新采样 sales
```

```python
print('月份最高的销量：', sales.max())
print('销量最高的月份：', sales.idxmax())

print('月环比的增长率：',sales.pct_change(),sep='\n')
```

```python
# 找出连续增长超过2个月的月份
a = sales.pct_change()
b = a > 0
print(b)
# b.rolling(3)
# b.rolling(3).sum()
print(b.rolling(3).sum() == 3)
print(b[b.rolling(3).sum() == 3].keys().tolist(),sep='\n')
```
