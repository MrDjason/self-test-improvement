# 一、ndarray
## 1.ndarry特性
多维性

```python
import numpy as np
```

创建一个零维数组

```python
arr = np.array(5) # 创建0维ndarry数组
print(arr)
print('arr的维度：', arr.ndim) # arr.ndim = number of dimensions arr
print(arr.dtype)
```

创建一个一维数组

```python
arr = np.array([1,2,3]) # 创建1维ndarry数组
print(arr)
print('arr的维度：', arr.ndim) # arr.ndim = arr.number of dimensions 
```

创建一个二维数组

```python
arr = np.array([[1,2,3],[4,5,6]]) # 创建2维ndarry数组
print(arr)
print('arr的维度：', arr.ndim)
```

5.索引与切片

```python
# 一维 数组的索引与切片
arr = np.random.randint(1, 100, 20)
```

```python
print(arr)
```

```python
print(arr[0])
print(arr[:3])
print(arr[(arr>10)&(arr<70)]) # 通过arr>10并且arr<70生成布尔掩码数组，通过arr[mask]用掩码筛选元素
# 布尔掩码数组 [False  True False  True False  True]
# Numpy只保留mask为True位置的元素
```

```python
# 二维数组的索引与切片
arr = np.random.randint(1,100,(4,8))
print(arr)
```

```python
print(arr[0,3]) # 获得34
print(arr[:2,:])
print() # 空行
print(arr[arr>50]) # 返回一维数组
print(arr[2]) # 获取第三行
print(arr[:,3])
```

# 标准差
$\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2}$
# 方差
$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 $

```python
# 计算标准差、方差
print(np.var([1,2,3]))
print(np.std([1,2,3]))

arr1 = np.array([1,2,1,2,1,1,1,2])
arr2 = np.array([1,0,3,0,0,0,4,3])
print(np.mean(arr1))
print(np.mean(arr2))

print(np.var(arr1))
print(np.var(arr2))
```

```python
# 计算最大值、最小值
arr = np.random.randint(1,200,8)
print(arr)
print(np.max(arr))
print(np.min(arr))
print(np.argmax(arr), np.argmin(arr))
```

```python
# 分位数
# 第 p 百分位数（p 取值 0~100）的严格定义：将数据从小到大排序后，满足两个条件的数值：
# 至少 p% 的数据小于等于该值；
# 至少 (100-p)% 的数据大于等于该值。

# 中位数
print(np.median([1,2,3]))
np.random.seed(0)
arr = np.random.randint(0,100,4)
print(arr)
print(np.median(arr))
print(np.percentile(arr,50)) # 取数列50%处的值，即中位值
print(np.percentile(arr,25))
# 1.取值→(n-1) * p / 100,n是数据个数，p是百分位【第二个位置的值，这里即25】
# 2.结果：索引整数部分值 + 小数部分 × （下一个索引的数值 - 索引整数部分的数值）
```

```python
# 累计和、累积积

arr = np.array([1,2,3])
print(np.sum(arr))
print(np.cumsum(arr))

print(np.cumprod(arr)) # 数组数值对应第x个对应索引数组的前x个数的乘积
```

# 比较函数  
比较是否大于、小于、等于  
逻辑与、或、非  
检查数组中是否有一个True，是否所有的都为True，自定义条件  

```python
# 是否大于
print(np.greater([3,4,5,6,7], 4))
# 是否小于
print(np.less([3,4,5,6,7,8],4))
# 是否等于
print(np.equal([3,4,5,6,7,8],4))
print(np.equal([3,4,5],[4,4,4]))
```

```python
print(np.logical_and([1,0],[1,1]))
print(np.logical_or([1,0],[1,1]))
print(np.logical_not([1,0]))
```

```python
# 检查元素是否至少有一个元素为True
print(np.any([0,0,0,0,0]))
# 检查是否全部元素为True
print(np.all([1,1]))
```

```python
# 自定义条件
# print(np.where(条件, 符合条件设置为..., 不符合条件设置为...))
arr = np.array([1,2,3,4,5])
print(np.where(arr>3,arr,0))
```

```python
arr = np.array([1,2,3,4,5])
print(np.where(arr<3,1,0))
```

```python
score = np.random.randint(50, 100, 20)
print(score)
print(np.where(score>=60, '及格', '不及格'))
print(np.where(
    score<60,'不及格',np.where(
        score<90,'良好','优秀'
    )
))
```

```python
# np.select(条件,返回的结果)
print(np.select([score>90,(score>=60)&(score<90),score<60],['优秀','良好','不及格'],default='未知'))
```

```python
# 排序函数
np.random.seed(0)
arr = np.random.randint(1,100,20)
print(arr)
# arr.sort()改变数列位置
print(np.sort(arr)) # 不改变位置排序
print(np.argsort(arr)) # 打印索引
print(arr)
```

```python
# 去重函数
print(np.unique(arr))
```

```python
# 数组拼接
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print(arr1+arr2)
print(np.concatenate((arr1,arr2)))
```

```python
# 数组的分割
print(np.split(arr, 5)) # 能等分的数组
```
