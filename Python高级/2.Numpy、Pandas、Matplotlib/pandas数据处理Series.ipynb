# Pandas

## Series

### Series的创建

```python
import pandas as pd

s = pd.Series([10,2,3,4,5])
print(s)
# 第一列 索引 第一列以后是值

# 自定义索引
s = pd.Series([10,2,3,4,5], index=['A','B','C','D','E'])
print(s)

# 定义name
s = pd.Series([10,2,3,4,5], index=['A','B', 'C', 'D', 'E'], name = '月份')
print(s)
```

```python
# 通过字典创建
# s = pd.Series([10,2,3,4,5], index=['A','B','C','D','E'])
s = pd.Series({'a':1, 'b':2, 'c':3, 'd':4, 'e':5})
print(s)
# 在s基础上取一个index获取一个新的Series 
s1 = pd.Series(s, index=['a','c'])
print(s1)
```

### Series的属性

```python
print(s.index)
print(s.values)
print(s.shape, s.ndim, s.size, sep='\n')
s.name = 'test'
print(s.dtype, s.name)
print(s)
```

```python
print(s.loc['b']) # 显示索引 location
print(s.iloc[1])  # 隐式索引

print(s.iloc[0:2])
print(s.at['a']) # at不支持切片
print(s.iat[0])
```

```python
# 访问数据
# print(s[1]) # 1标签还是位置，python搞不清
print(s['a'])
print(s[s<3])
s['f'] = 6 #增加一行
print(s)
print(s.head()) # s.head()获取数据最前N行，默认只打印五行

print(s.tail()) # s.head()获取数据最后N行，默认只打印五行
```

### Series的常见函数

```python
import numpy as np
s = pd.Series([10,2,np.nan,None,3,4,5], index=['A', 'B', 'C', 'D', 'E', 'F', 'G'], name='Data')
print(s)
```

```python
s.head(3) # 取前3行
s.tail(2) # 取后2行
```

```python
# 查看所有的描述性信息
s.describe() # 忽略缺失值
```

```python
# 获取元素个数(忽略缺失值)
print(s.count())
print(s.keys()) # 方法
print(s.index)  # 属性
```

```python
print(s.isna()) # 检查Series里的每一个元素是否为缺失值
s.isin([4,5,6])
```

```python
print(s)
print(s.mean())
print(s.std())
print(s.var())
print(s.min())
print(s.max())
print(s.median())
```

```python
print(s.sort_values())
print(s.quantile(0.8))
# ---------------------
# 2   3   4   5   10
# 4（四段）*0.8=3.2
# 比第三段（4-5）大，所以比5大，比10小
# 5 + （10-5）* 0.2 = 6
```

```python
# 众数 
print(s.mode()) # 出现频率一样

s['H'] = 4
print(s.mode()) # 此时显示4，出现次数最多
```

```python
print(s.value_counts())
```

```python
# 去重
print(s.drop_duplicates())
print(s.unique())
print(s.nunique()) # 去重后的元素个数
```

```python
# 排序、值、索引
print(s.sort_index())  # 按索引排序
print(s.sort_values()) # 按值排序
```
