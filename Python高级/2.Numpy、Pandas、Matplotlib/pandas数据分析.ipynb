### 缺失值处理

```python
import pandas as pd
import numpy as np
s = pd.Series([1, 2, np.nan, None, pd.NA])
df = pd.DataFrame([[1,pd.NA,2],[2,3,5],[None,4,6]])
print(s)
print('-'*20)
# 查看是否是缺失值
print(s.isna())
print('-'*20)

print(df)
print('-'*20)
print(df.isnull())

print(s.isna().sum(),df.isnull().sum(),sep='\n\n')
```

```python
# 剔除缺失值
print(s)
print('-'*30)
print(s.dropna()) # 剔除缺失值那一行
print('-'*30)

print(df)
print('-'*30)
print(df.dropna(how='all')) # 所有值都是缺失值，删除这一行
print('-'*30)
print(df.dropna(thresh=3)) # 如果至少有n个值不是缺失值，就保留
print(df.dropna(axis=1)) # 剔除一整列的记录
# print(df.dropna(subset=['第1列'])) # 如果某列有缺失值则删除这一行
```

```python
# 填充缺失值
file_path_2 = os.path.join(os.path.dirname(file_path), 'weather_withna.csv')
df = pd.read_csv(file_path_2)
print(df.isna().sum(axis=0))
print(df.head())
print('-'*30)
print(df.fillna({'temp_max':20, 'wind':2.5}).tail()) # 使用字典填充后（看tail）
print(df.fillna(df[['wind']].mean()).tail()) # 使用统计值来填充 
```

```python
print(df)
print('-'*30)
print(df.ffill().tail()) # 用前面数据添加 Ffill(F:font)
print(df.bfill().tail()) # 
```

### 重复数据处理

```python
print(df)
print('-'*30)
print(df.duplicated()) # 一整条记录都是一样的，标记为重复，返回True
print(df.drop_duplicates()) # 删除重复的 
print(df.drop_duplicates(subset=['name'])) # 根据指定列去重
df.drop_duplicates(subset=['name'], keep='last') # 保留最新出现的行
```
