# Matplotlib学习

- 折线图 plot   : 趋势随时间变化
- 条形图 bar    :类别之间的对比
- 饼图   pie    :整体组成比例
- 散点图 scatter:两变量相关性
- 箱线图 boxplot:数据分布、异常

### 折线图 plot

```python
# 绘制、创建、显示
# 1.绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei' # win:SimHei

# 2.创建图表，设置大小
plt.figure(figsize=(10,5))

# 要绘图的数据
month = ['1月', '2月', '3月', '4月']
sales = [100, 150, 80, 130]
# 绘制折线图
plt.plot(month, sales,
        label='产品A',   # plt.legend()在绘制的时候配置
        color='orange', # 颜色
        linewidth=2,    # 线宽 
        linestyle='--', # 虚线折线
        marker='o')     # 转折点变为实心圆圈

# 添加标题
plt.title('2025年销售趋势', color='red', fontsize=20)

# 添加坐标轴标签
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额(万元)', fontsize=10)

# 添加图例
plt.legend(loc='upper left')

# 添加网格线
plt.grid(True, alpha=0.1, color='blue', linestyle='--')

# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 设置y轴的范围
plt.ylim(0,160)

# 在每个数据点上显示数值
for x,y in zip(month, sales):
    print(x, y)
    plt.text(x,y+5,str(y),ha='center',va='bottom', fontsize=10) # 将y标签的数值往上提5

# 3.显示图表
plt.show()
```

### 柱状图 bar 条形图 barh

```python
# 要绘制的数据
subjects = ['语文', '数学', '英语', '科学']
scores = [85, 92, 78, 88]
# 添加标题
plt.title('2025年成绩分布',
          color='red',
          fontsize=20)
# 添加坐标轴标签
plt.xlabel('科目', fontsize=12)
plt.ylabel('分数', fontsize=12)

# 绘制柱状图
plt.bar(subjects, scores,
        label='小明',
        color='orange',
        width=0.4,)

# 添加图例
plt.legend(loc='upper left') # 图例必须在绘制下方

# 添加网格线
plt.grid(axis='y',alpha=0.1,color='blue', linestyle='--')

# 设置y轴范围
plt.ylim(0,100)

# 设置刻度字体大小
plt.xticks(rotation=0, fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 数据点上显示
for x,y in zip(subjects, scores):
    plt.text(x,y+1,str(y), ha='center', va='bottom', fontsize=12)

# 自动优化排版
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10,5))

subjects = ['语文Chinese', '数学math', '英语English', '科学Science']
scores = [85, 92, 78, 88]

plt.barh(subjects, scores)
plt.title('科目', color='red', fontsize=20)
plt.xlabel('分数', fontsize=12)
plt.ylabel('科目', fontsize=12)
plt.show()
# 底部文字规整用柱状图，否则用条形图
```

### 饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'
plt.figure(figsize=(10,5))

# 要绘图的数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]

# 绘制饼图 沿一至四绘制
plt.pie(times, labels=things,
        autopct='%1.1f%%',
        startangle=90)

# 添加标题
plt.title('一天时间分布', color='red', fontsize=20)

# 自动布局
plt.tight_layout()
# 显示图标
plt.show()
```

### 环形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'
plt.figure(figsize=(10,5))

# 要绘图的数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]

# 绘制饼图 沿一至四绘制
colors = ['#839096','#b4c7dc', '#d4b4a0', '#c7d9b7', '#e5c8e0']
plt.pie(times, labels=things,
        autopct='%1.1f%%', # 显示百分比
        startangle=90, # 调整初始画图的角度
        colors = colors, # 设置饼图配色
        wedgeprops={'width':0.6}, # 设置圆环宽度
        pctdistance=0.7, # 设置百分比的位置
        )  

# 添加标题
plt.title('一天时间分布', color='red', fontsize=20)
plt.text(0,0,'总计100%%')
# 自动布局
plt.tight_layout()
# 显示图标
plt.show()
```

### 爆炸式饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'SimHei'
plt.figure(figsize=(10,5))

# 要绘图的数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]

# 爆炸式
explode = [0.1,0,0,0,0] # 设置突出块的位置

# 绘制饼图 沿一至四绘制
plt.pie(times, labels=things,
        autopct='%1.1f%%',
        explode = explode,
        shadow=True,
        )

# 添加标题
plt.title('一天时间分布', color='red', fontsize=20)

# 自动布局
plt.tight_layout()
# 显示图标
plt.show()
```

```python
# 散点图
import matplotlib.pyplot as plt
from matplotlib import rcParams
import random

rcParams['font.family'] = 'SimHei'

plt.figure(figsize=(10,8))

# 添加标题
plt.title('X变量与Y变量的关系', color='red', fontsize=20)

# 要绘图的数据
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0, 10)
    x.append(tmp)
    y.append(2*tmp + random.gauss(0,2))
 
# 绘制散点图
plt.scatter(x, y,
            color='blue',
            alpha=0.5,
            s=20,
            label='数据')

# 添加图例 
plt.legend(loc='upper left')

# 添加坐标轴标签
plt.xlabel('X自变量', fontsize=10)
plt.ylabel('Y变量', fontsize=10)

# 设置y轴的范围
plt.ylim(0, 30)
plt.plot([0,10],[0,20],color='red',linewidth=2,linestyle='-')

# 添加网格线
plt.grid(True, alpha=0.1, color='blue', linestyle='--')

plt.show()
```

### 箱线图

```python
import matplotlib.pyplot as plt

data = {
    '语文':[82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学':[75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语':[70, 72, 68, 65, 78, 80, 85, 90, 95]
}

plt.figure(figsize=(8,6))
plt.boxplot(data.values(), tick_labels=data.keys())

plt.title('各科成绩分布(箱线图)')
plt.ylabel('分数')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```
