# 3.贝尔曼方程

## 3.1 贝尔曼方程的推导

### 3.1.1 概率和期望值
$$
\mathbb{E}[x]=\sum_xxp(x)
$$

$$
\begin{aligned}
\mathbb{E}[r(x,y)] &=\sum_x \sum_y p(x,y)r(x,y) \\
&=\sum_x \sum_y p(x)p(y|x)r(x,y)
\end{aligned}
$$

### 3.1.2 贝尔曼方程的推导
回顾对收益的定义  
$$
G_t = R_t + \gamma R_{t+1} + \gamma ^2R_{t+2} + ...
$$

将t+1带入式3.1中的t中
$$
G_{t+1} = R_{t+1} + \gamma R_{t+2} + \gamma ^2R_{t+3} + ...
$$
变形  
$$
\begin{aligned}
G_t &= R_t + \gamma R_{t+1} + \gamma ^2R_{t+2} + ... \\
    &= R_t + \gamma(R_{t+1} + \gamma R_{t+2} + ...) \\
    &= R_t + \gamma G_{t+1}
\end{aligned}
$$

状态价值函数数学式如下
$$
v_\pi(s)=\mathbb{E}_\pi[G_t|S_t=s]
$$
将$G_t$带入其中
$$
\begin{aligned}
v_\pi(s) &=\mathbb{E}_\pi[R_t + \gamma G_{t+1}|S_t=s] \\
&= \mathbb{E}_\pi[R_t|S_t=s] + \gamma \mathbb{E}_\pi[G_{t+1}|S_t=s]
\end{aligned}
$$

其中期望 
 
---
$$
\mathbb{E}_\pi [R_t|S_t=s]
$$
首先
- $R_t=r(s,a,s^\prime)$
- $R_t$ 由动作 $a$ 以及下一状态 $s^\prime$ 决定

由**条件期望**的定义 
$$
\mathbb{E}[X|C] = \sum\limits_xx·P(X=x|C)
$$
$$
\mathbb{E}[f(X,Y)|A]=\sum\limits_x \sum\limits_y f(x,y)·P(X=x,Y=y|A)
$$
得
$$
\mathbb{E}_\pi[R_t|S_t=s]=\sum\limits_r R_t·P(R_t=r|S_t=s)=\sum\limits_r r·P(R_t=r|S_t=s)
$$
由**全概率公式**的定义
$$
P(X|C) = \sum\limits_i P(X,B_i|C)
$$
得
$$
P(R_t=r|S_t=s)=\sum\limits_a\sum\limits_{s^\prime}P(R_t=r,A_t=a,S_{t+1}=s^\prime|S_t=s)
$$
再简化
$$
P(r,a,s^\prime|s) = P(r|s,a,s^\prime) · P(a,s^\prime|s)
$$
因为在 $MDP$ 中，$R_t$ 只由 $s,a,s^\prime$ 决定，当$s,a,s^\prime$ 被确立下来，则 $R_t$​ 被唯一确定为常数 $r$，此时事件$P(R_t=r|s,a,s^\prime)=1$  
所以
$$
P(r,a,s^\prime|s) = P(a,s^\prime|s)
$$
且因为


因此
$$
\begin{align}
\mathbb{E}_\pi[R_t|S_t=s] &=\sum\limits_a\sum\limits_{s^\prime}\sum\limits_r r · P(r,a,s^\prime|s) \\
&= \sum\limits_a\sum\limits_{s^\prime} P(a,s^\prime|s) \sum\limits_r r · P(r|s,a,s^\prime) \\
&= \sum\limits_a\sum\limits_{s^\prime} P(a,s^\prime|s) [r(s,a,s^\prime)·1 + \sum_{r≠(s,a,s^\prime)} r·0] \\
&= \sum\limits_a\sum\limits_{s^\prime} P(a,s^\prime|s) · r(s,a,s^\prime)
\end{align}
$$

其中，现在状态 $s$ 以概率 $\pi(a|s)$ 通过动作 $a$ 以概率 $p(s^\prime|s,a)$ 到达下一状态$s^\prime$（需要根据图来推导）  
所以  
$$
\mathbb{E}_\pi [R_t|S_t=s]=\sum_a\sum_{s^\prime}P(a,s^\prime|s)r(s,a,s^\prime)=\sum_a\sum_{s^\prime}\pi(a|s)p(s^\prime|s,a)r(s,a,s^\prime)\Rightarrow\sum_{a,s^\prime} \pi(a|s)p(s^\prime|s,a)r(s,a,s^\prime)
$$
---

$$
\gamma \mathbb{E_\pi}[G_{t+1}|S_t=s]
$$

剩下的这一项中 $\gamma$ 是常数，注意到与状态价值函数定义式相似，但 $G_{t+1}$ 部分有所不同  
由状态价值函数 $v_\pi(s)=\mathbb{E}_\pi[G_t|S_t=s]$ ，将 $t+1$ 带入
$$
v_\pi(s)=\mathbb{E}_\pi[G_{t+1}|S_{t+1}=s]
$$
现在需要解决 $S_{t+1}=s$ 时的价值函数，关键在于将条件 $S_t=s$ 变为 $S_{t+1}=s$ 的形式，换句话来说就是进入下一个时刻


首先，我们希望得到 $E_\pi [G_{t+1}|S_t=s]$。
- 由全期望公式 $\mathbb{E}[X]=\sum_y \mathbb{P}(Y=y)·\mathbb{E}(X|Y=y)$
$$
\boldsymbol{
E_\pi [G_{t+1}|S_t=s] = \sum_a \pi(a|s)\mathbb{E}[G_{t+1}|s,a]
}
$$
其中 
$$
\mathbb{E}[G_{t+1}|s, a] = \sum_{s^\prime}p(s^\prime|s,a)·\mathbb{E}[G_{t+1}|s,a,s^\prime]
$$  

- 因为马尔可夫性，未来只取决于当前状态，因此 $S_t=s,A_t=a$ 忽略。再代入 $v_\pi(s)=\mathbb{E}_\pi[G_{t+1}|S_{t+1}=s]$


$$
\sum_{s^\prime}p(s^\prime|s,a)·\mathbb{E}[G_{t+1}|s,a,s^\prime]=\sum_{s^\prime}p(s^\prime|s,a)·\mathbb{E}[G_{t+1}|S_{t+1}=s^\prime]=\sum_{s^\prime}p(s^\prime|s,a)v_\pi(s^\prime)
$$

合并公式
$$
\mathbb{E}[G_{t+1}|S_t=s] = \sum_{a,s^\prime}\pi(a|s)p(s^\prime|s,a)v_\pi(s^\prime)
$$
---

因此可得

$$
\begin{align}
v_\pi(s) &= E_\pi[R_t|S_t=s] + \gamma E_\pi[G_{t+1}|S_t=s] \nonumber\\
&= \sum_{a,s^\prime}\pi(a|s)p(s^\prime|s,a)r(s,a,s^\prime) + \gamma\sum_{a,s^\prime}\pi(a|s)p(s^\prime|s,a)v_\pi(s^\prime) \nonumber\\
&= \sum_{a,s^\prime}\pi(a|s)p(s^\prime|s,a)[r(s,a,s^\prime)+\gamma v_\pi(s^\prime)] \tag{1-1}
\end{align}
$$
(1-1)就是贝尔曼公式。贝尔曼方程是表示状态s的价值函数和下一个可能的状态 $s^\prime$ 的价值函数之间关系的式子

## 3.3 行动价值函数与贝尔曼方程

### 3.3.1 行动价值函数
先看一下状态价值函数  
$$
v_\pi(s) = \mathbb{E}_\pi[G_t|S_t=s]
$$
状态价值函数有两个条件，分别是状态 $s$ 和 策略 $\pi$ ，在这两个条件的基础上，可以考虑增加一个额外条件：行动 $a$
$$
q_\pi(s,a)=\mathbb{E}\pi[G_t|S_t=s,A_t=a]
$$
$Q$ 函数在时刻 $t$ 的状态下 $s$ 下采取行动 $a$，并从时刻 $t+1$ 开始根据策略 $\pi$ 采取行动，此时得到的收益的期望值是 $q_\pi(s,a)$  
$Q$ 函数是在状态价值函数的基础上，将行动a增加为条件的函数——原本状态价值函数 $v_\pi(s)$ 是通过策略 $\sum_a\pi(a|s)$ 到达 $a$ 后继续，现在 $Q$ 函数是通过将行动 $a$ 纳为条件后继续。如果想办法使得 $Q$ 函数中根据策略 $\pi$ 来选择行动 $a$ ，那么 $Q$ 函数和状态价值函数就一样了。  

假设在状态 $s$ 中有 ${a_1,a_2,a_3}$ 三个后算行动，那么此时将根据策略 $\pi$ 行动

- 选择行动 $a_1$ 的概率为 $\pi(a_1|s)$ ，此时 $Q$ 函数为 $q_\pi(s,a_1)$
- 选择行动 $a_2$ 的概率为 $\pi(a_2|s)$ ，此时 $Q$ 函数为 $q_\pi(s,a_2)$
- 选择行动 $a_3$ 的概率为 $\pi(a_3|s)$ ，此时 $Q$ 函数为 $q_\pi(s,a_3)$

这种情况下，收益的期望值可作为 $Q$ 函数的加权和来计算，数学式如下：
$$
\pi(a_1|s)q_\pi(s,a_1) + \pi(a_2|s)q_\pi(s,a_2) + \pi(a_3|s)q_\pi(s,a_3) = \sum_{a=\{a_1,a_2,a_3\}}\pi(a|s)q_\pi(s,a)
$$
$$
v_\pi(s) = \sum_a\pi(a|s)q_\pi(s,a)
$$

### 3.3.2 使用行动价值函数的贝尔曼方程
推导行动价值函数（$Q$函数）的贝尔曼方程  
$$
\begin{align}
q_\pi(s,a) &= \mathbb{E}[G_t|S_t=s,A_t=a] \\
&= \mathbb{E}[R_t+\gamma G_{t+1}|S_t=s,A_t=a]
\end{align}
$$
这里，行动 $a$ 和状态 $s$ 都是固定的，这种情况下，以 $p(s^\prime|s,a)$ 的概率迁移到下一个状态 $s^\prime$ ，奖励由函数 $r(s,a,s^\prime)$ 给出  
- 注：环境里没有直接给出 $p(r|s,a)$ ，但是我们能通过以上方法轻松获取 $p(s^\prime|s,a)$ ，所以我们要用全概率公式将对 $r$ 的求和，替换为对 $s^\prime$ 的求和
- 把所有能给 r 的状态概率加起来，就是拿 r 的概率 $p(r|s,a) = \sum_{\substack{s^\prime \\ r(s,a,s^\prime)=r}}p(s^\prime|s,a)$
- 带入期望公式
$$
\begin{align}
\mathbb{E}[R_t | s,a] &= \sum_{r} r \cdot \left( \sum_{\substack{s' \\ r(s,a,s')=r}} p(s' | s,a) \right) \\
&= \sum_{s'} p(s' | s,a) \cdot r(s,a,s')
\end{align}
$$

则

$$
\begin{align}
q_\pi(s,a) &= \mathbb{E}_\pi[R_t + \gamma G_{t+1}|S_t=s,A_t=a] \\
&= \mathbb{E}_\pi[R_t|S_t=s,A_t=a] + \gamma\mathbb{E}_\pi[G_{t+1}|s,a] \\
&= \sum_{s^\prime} r(s,a,s^\prime) · p(s' | s,a) + \gamma \sum_{s'}p(s'|s,a)\mathbb{E}_\pi(G_{t+1}|s,a,s')\\
&= \sum_{s^\prime} p(s' | s,a) · r(s,a,s^\prime)  + \gamma \sum_{s'}p(s'|s,a)\mathbb{E}_\pi(G_{t+1}|S_{t+1}=s')\\
&=\sum_{s^\prime}p(s^\prime|s,a)[r(s,a,s^\prime) + \gamma v_\pi(s^\prime)]
\end{align}
$$

将 $v_\pi(s)=\sum\limits_a \pi(a|s)q_\pi(s,a)$ 带入 $v_\pi(s^\prime)$  
$$
q_\pi(s,a) = \sum_{s^\prime}p(s^\prime|s,a)\left\{r(s,a,s^\prime) + \gamma \sum\limits_{a'} \pi(a'|s')q_\pi(s',a')\right\}
$$
此时， $a'$ 表示在时刻 $t+1$ 采取的行动，此式就是使用 $Q$ 函数的贝尔曼方程

## 3.4 贝尔曼最优方程

### 3.4.1 状态价值函数的贝尔曼最优方程

首先从贝尔曼方式开始

$$
\begin{align}
v_\pi(s) &= \sum_{a,s'} \pi(a|s)p(s'|s,a)[r(s,a,s') + \gamma v_\pi(s')] \\
&=\sum_a \pi(a|s) \sum_s p(s'|s,a) \left\{ r(s,a,s') + \gamma v_\pi(s')\right\}
\end{align}
$$

最优策略如果是 $\pi_*(a|s)$ ，则有以下贝尔曼方程成立  
$$
v_*(s) = \sum_a \pi_*(a|s) \sum_{s'}p(s'|s,a)\left\{r(s,a,s') + \gamma v_*(s')\right\}
$$
$\sum_a \pi_*(a|s)$ 是在当前状态选择每个行动的概率，也是分配动作价值的权重

最优策略的目标是最大化状态价值函数，可取以下式子作为**贝尔曼最优方程**
$$
v_*(x) = \max_a \sum_{s'}p(s'|s,a)\left\{r(s,a,s') + \gamma v_*(s')\right\}
$$



### 3.4.2 Q函数的贝尔曼最优方程

以下是Q函数的贝尔曼方程
$$
q_\pi(s,a) = \sum_{s^\prime}p(s^\prime|s,a)\left\{r(s,a,s^\prime) + \gamma \sum\limits_{a'} \pi(a'|s')q_\pi(s',a')\right\}
$$
带入最优策略 $\pi_*$
$$
q_*(s,a) = \sum_{s^\prime}p(s^\prime|s,a)\left\{r(s,a,s^\prime) + \gamma \sum\limits_{a'} \pi_*(a'|s')q_*(s',a')\right\}
$$
同样可以使用相同方法获得 $Q$ 函数的贝尔曼最优方程——
$$
q_*(s,a) = \sum_{s^\prime}p(s^\prime|s,a)\left\{r(s,a,s^\prime) + \gamma \max_{a'}q_*(s',a')\right\}
$$

## 3.5 贝尔曼最优方程的示例

### 3.5.1 应用贝尔曼最优方程
$$
v_*(s) = \max_a \sum_{s'}p(s'|s,a)\left\{r(s,a,s') + \gamma v_*(s')\right\}
$$
如果状态迁移是确定性的，则 $s'=f(s,a)$  
- $s'=f(s,a)$，则 $p(s'|s,a)=1$  
- $s'≠f(s,a)$，则 $p(s'|s,a)=0$  
所以得
$$
v_*(s) = \max_a \left\{r(s,a,s') + \gamma v_*(s')\right\}
$$

## 3.6 小结

**贝尔曼方程**  
$$
v_\pi(s) = \sum_{a,s'} \pi(a|s) p(s'|s,a) \left\{ r(s,a,s') + \gamma v_\pi(s') \right\}
$$

$$
q_\pi(s,a) = \sum_{s'} p(s'|s,a) \left\{ r(s,a,s') + \gamma \sum_{a'} \pi(a'|s') q_\pi(s',a') \right\}
$$

**贝尔曼最优方程**

$$
v_*(s) = \max_a \sum_{s'} p(s'|s,a) \left\{ r(s,a,s') + \gamma v_*(s') \right\}
$$

$$
q_*(s,a) = \sum_{s'} p(s'|s,a) \left\{ r(s,a,s') + \gamma \max_{a'} q_*(s',a') \right\}
$$

**最优策略**

$\arg\max\limits_a q_*(s,a)$ 在所有可选动作 $a$ 中，找到让 $q_*(s,a)$ 最大的那个 $a$

$$
\begin{align}
\mu_*(s) &= \underset{a}{\arg\max} \, q_*(s,a) \\
&= \underset{a}{\arg\max} \sum_{s'} p(s'|s,a) \left\{ r(s,a,s') + \gamma v_*(s') \right\}
\end{align}
$$
