## 2.2环境和智能代理的数学表示

### 2.2.1 状态迁移

状态迁移函数  

$$
s^\prime = f(s,a)
$$

- $s^\prime$ 下一状态
- $s$ 输入状态
- $a$ 行动

迁移到下一个状态$s^\prime$的概率可以用**状态迁移概率**表示
$$
p(s^\prime|s,a)
$$

$p(s^\prime|s,a)$决定了下一个状态$s^\prime$只取决于当前状态$s$和行动$a$  
状态迁移不需要过去的信息，这个特性被称之为马尔可夫性  

### 2.2.2 奖励函数
**奖励函数**
$$
r(s,a,s^\prime)
$$
- $r$ 奖励函数
- $s$ 输入状态
- $a$ 行动
- $s^\prime$ 下一状态

如 $r(s=L2,a=Left,s^\prime=L1)=1$ 代表此时身处L2位置，即将向左走，下一个状态位置在L1处，这个动作的奖励为1

### 2.2.3 智能代理的策略
智能代理的行动可以是**确定性**的，也可以是**随机性**的

**确定性的策略**在某个状态下总是采取固定的行动，如智能代理在某一位置时一直采取向左移动的行动，这样确定性的策略可以定义为如下所示的函数
$$
a = \mu(s)
$$
$\mu(s)$是以给定的状态作为参数并返回行动$a$的函数

**随机性的策略**依概率行动，数学公式如下
$$
\pi(a|s) \Leftarrow p(a|s)
$$
表示在为状态$s$下采取行动$a$的概率。特地使用 $\pi$ 是为了区分智能体策略

## 2.3 MDP的目标
MDP的目标是找到**最优策略**，最优策略是使收益最大化的策略

### 2.3.1 回合制任务和连续性任务
围棋被视为回合制任务。回合制任务中的回合的意思是从开始到结束的一系列操作  
连续性任务则是没有结束的问题，例如库存管理问题

### 2.3.2 收益
收益(return)，智能代理的目标是使收益最大化  
设时刻为$t$，状态为$S_t$，智能代理根据策略$\pi$执行行动$A_t$，获得奖励$R_t$，之后迁移到新状态$S_{t+1}$，这个过程不断重复进行。这种情况下收益$G_t$的定义如下
$$
G_t = R_t + \gamma R_{t+1} + \gamma ^2R_{t+2} + ...
$$
- $G_t$表示智能代理获得的奖励之和
- $\gamma$ 折现率，被设定为0.0和1.0之间的实数。引入折现率主要防止连续性任务的收益变为无穷大

### 2.3.3 状态价值函数
智能代理和环境的行动可能是随机性的，智能代理可能随机地决定行动，状态也可能随机迁移，这个情况下得到的收益可能呈现随机的特点  
即相同的状态开始，不同回合的收益也随机变化。为了处理这种随机行动，需要引入收益的期望值作为衡量标准。**收益的期望值**数学公式如下：
$$
v_\pi(s)=\mathbb{E}[G_t|S_t=s,\pi]
$$

- $S_t=s,\pi$ $S_t$状态为s，智能代理策略为$\pi$
- $v_\pi(s)$ 收益的期望值，就是**状态价值函数**
  - $v_\pi(s)$ 将$\pi$写在$v$右下角的原因是智能代理的策略被作为条件给出，策略$\pi$发生变化，智能代理获得的奖励也会发生变化，收益也会变化

也可写作
$$
v_\pi(s)=\mathbb{E}_\pi[G_t|S_t=s]
$$

### 2.3.4 最优策略和最优价值函数
如何判断最优策略？  
假如有两个状态价值函数 $v_\pi(s)$ 与 $v_{\pi^\prime}(s)$，在所有状态下，$v_{\pi^\prime}(s)≥v_\pi(s)$，就可以说 $\pi^\prime$ 是比 $\pi$ 更好的策略  
如果用 $\pi_*$ 表示最优策略，那么策略 $\pi_*$ 就是与其他策略相比，在所有状态下的状态价值函数 $v_{\pi_*}(s)$ 的值最大的策略

MDP中至少存在一个最优策略，此外最优策略是确定性策略，确定性策略中每个状态下的行动是唯一确定的，用数学式可以表示为函数 $\mu_*$ ，有 $a=\mu_*(s)$ 成立，当输入状态 $s$ 时，会输出行动 $a$
